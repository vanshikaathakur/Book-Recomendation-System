##Population Based Recommendation System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings'] >= 250] \
    .sort_values('avg_ratings', ascending=False) \
    .head(50)
popular_df
```

```python
popular_df=popular_df.merge(books, on='Book-Title') \
    .drop_duplicates('Book-Title') \
    [['Book-Title', 'Book-Author', 'Image-URL-S', 'num_ratings', 'avg_ratings']]
```

## Colaborative Filtering Based Recommender System

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
educated_users=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(educated_users)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_rating
```

```python
# final_rating = final_rating.reset_index()
```

```python
final_rating.columns
```

```python
pt=final_rating.pivot_table(
    index='Book-Title',
    columns='User-ID',
    values='Book-Rating'
)
```

```python
pt.fillna(0, axis=1, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
similarity_score
```

```python
#5 similar books to the chosen
sorted(list(enumerate(similarity_score[0])),key=lambda x:x[1],reverse=True) [1:6]           #distance with index,sorting on the bases of similarity score
```

```python
# Assuming you want to compute similar_items inside the function
def recommend(book_name): # Add the necessary variables as arguments
    index = np.where(pt.index == book_name)[0][0]

    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True) [1:6]
    data=[]
    # Print the recommendations
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-S'].values))

        data.append(item)
    return data
```

```python
recommend('A Walk to Remember')
```
